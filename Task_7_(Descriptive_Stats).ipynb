# Median

Median is the middle value after sorting the data.In medain if we want to find the middle value we need to check two conditions

1.number of elements in the data(even/odd)

2.Is the data  sorted or not


If the data is sorted and the no.of elements are odd  we can choose the middle element very easily.

median=$$\frac{(n+1)}{2}$$


if the no.of elements are even then we need to take the average of (n/2)th value and the ((n/2)+1) value

median=avg(n/2 ,((n/2)+1)



Median will not impacted by the the outliers


```python
#example of median
l=[1,2,3,4,5,6,7,8,9]#the elements are already are in sorted order with odd number of elements
a=len(l)+1
median=a/2
print(median)
```

# Variance
Variance will tell you how the data is spread around the mean

Population variance($\sigma^2$)=$$\frac{\sum_{i=1}^n ({x_i-\mu)}^2 }{N}$$


Sample variance(S)=$$\frac{\sum_{i=1}^n ({x_i-\mu})^2} {(N-1)}$$

A large variance indicates that the data is spread out,  a small variance indicates that the data is clustered closely around the mean.




```python
#variance without library function
data['Annual_HH_Income'].var()
```

```python
#variance with library function 
print("the variance is:",stats.variance(data['Annual_HH_Income']))
```

# Standard deviation
The standard deviation is a measure of the amount of variation or dispersion of a set of values.


A low standard deviation indicates that the values tend to be close to the mean (also called the expected value) of the set, while a high standard deviation indicates that the values are spread out over a wider range.
std($\sigma$)=$\sqrt[]{\sigma^2}$ 




standard deviation is the square root of variance

# Normal distribution

Normal Distribution, also called the Gaussian Distribution, is the most significant continuous probability distribution. Sometimes it is also called a bell curve.

A random variable  X is said to have a NormalDistribution, if its density fumction or probability distribution is given as 


f(x;$\mu$,$\sigma$)=$e^{-(x-\mu)^2/2\sigma^2}$/($\sigma$$\sqrt[]{2\pi}$)



where $\mu$=mean and $\sigma$=standard deviation are the two parameters of the normal distribution


A normal distribution is said to be symmetric because mean,median,mode will coincide

i.e.,mean=medain=mode

```python
#normal distribution
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from scipy import stats


sns.distplot(data['Annual_HH_Income'],hist=False)
```

# The area under normal curve is distributed as follows

1.Area of normal curve between $\mu$-$\sigma$ and $\mu$+$\sigma$ is 68.27%

p($\mu$-$\sigma$<x<$\mu$+$\sigma$)=0.6826

2.Area of normal curve between $\mu$-2$\sigma$ and $\mu$+2$\sigma$ is 95.43%


3.Area of normal curve between $\mu$-3$\sigma$ and $\mu$+3$\sigma$ is 99.7%





```python
one_std_right=mean_normal+(1*std_normal)
one_std_left=mean_normal-(1*std_normal)
two_std_right=mean_normal+(2*std_normal)
two_std_left=mean_normal-(2*std_normal)
three_std_right=mean_normal+(3*std_normal)
three_std_left=mean_normal-(3*std_normal)
print(one_std_right)
```

# Skewness

Positively skewed: Most frequent values are low and tail is towards high values.

Negatively skewed: Most frequent values are high and tail is towards low values.

If Mode< Median< Mean then the distribution is positively skewed.

If Mode> Median> Mean then the distribution is negatively skewed.


In statistics, skewness is a measure of the asymmetry of the probability distribution of a random variable about its mean. In other words, skewness tells you the amount and direction of skew (departure from horizontal symmetry). The skewness value can be positive or negative, or even undefined. If skewness is 0, the data are perfectly symmetrical, although it is quite unlikely for real-world data. As a general rule of thumb:

- If skewness is less than -1 or greater than 1, the distribution is highly skewed.
- If skewness is between -1 and -0.5 or between 0.5 and 1, the distribution is moderately skewed.
- If skewness is between -0.5 and 0.5, the distribution is approximately symmetric.

# Box - Cox Transformation


Box-Cox requires the input data to be strictlt positive(not even zero is acceptable)

For the features which have zeros or negative values Yeo-johnson comes to the rescue

```python
#Box-Cox transformation will help you to transform any distribution to a normal distribution
```

```python
def plot(df,var,t):
  plt.figure(figsize=(13,15))
  plt.subplot(1)
  sns.kdeplot(data[var])
  plt.title('before'+str(t).split('(')[0])

  plt.sunplot(2)
  p1=t.fit_transform(data[[var]]).flatten()
  sns.kdeplot(p1)
  plt.title('after'+str(t).split('(')[0])
```

```python
from sklearn.preprocessing import PowerTransformer
```

```python

def plots(df, var, t):
    plt.figure(figsize=(13,5))
    plt.subplot(121)
    sns.kdeplot(df[var])
    plt.title('before ' + str(t).split('(')[0])
    
    plt.subplot(122)
    p1 = t.fit_transform(df[[var]]).flatten()
    sns.kdeplot(p1)
    plt.title('after ' + str(t).split('(')[0])
```

```python
for col in data.columns:
    plots(data, col, PowerTransformer(method='box-cox'))
# The Box-Cox transformation can only be applied to strictly positive data
```
